[20] Determine if the Fourier Transform $X(\omega)$ of a signal $x[n]$ is a) even or odd, and b) imarinary or real when $x[n]$ is:
1. Real and even.
2. Imaginary and even.
3. Real and odd.
4. Imaginary and odd.

Show your work.

__Answer for 1.1:__

1. $x[n] = x_R[n]$, where $x_R[n]$ is the real part of $x[n]$. Equivalently, $x_I[n]=0$, where $x_I[n]$ is the imaginary part of $x[n]$.
2. We know that $$X(\omega) = \sum_{n=-\infty}^{\infty} \left(x_R[n]\cos(\omega n)+x_I[n]\sin(\omega n)\right)+j\left( -x_R[n]\sin(\omega n)+x_I[n]\cos(\omega n)\right).$$
3. The sum of an odd function from $-\infty$ to $\infty$ is zero. Thus, $$\sum_{n=-\infty}^{\infty}-x_R[n]\sin(\omega n)=0$$
4. Thus, $X(\omega) = \sum_{n=-\infty}^{\infty} x_R[n]\cos(\omega n).$ Thus, $X(\omega)$ is a sum of a real signal, hence it is always **real**.
5. Also, $X(-\omega) = \sum_{n=-\infty}^{\infty} x_R[n]\cos(-\omega n) = \sum_{n=-\infty}^{\infty} x_R[n]\cos(\omega n) = X(\omega)$, so it is **even**.
---

__Answer for 1.2:__

Now, we only have $x_I[n].$ so, following a similar analysis as 1.1 (as the $x_I[n]\sin(\omega n)$ is odd), we end up with $$X(\omega) = \sum_{n=-\infty}^{\infty} jx_I[n]\cos(\omega n)$$

1. Since this is the sum of purely imaginary numbers, $X(\omega)$ is **imaginary**.
2. Similar to 1.1, $X(\omega)= X(-\omega),$ hence it is **even**.

---

__Answer for 1.3:__

Now, since $x_R[n]$ is odd, $x_R[n]\sin(\omega n)$ is going to be even. Hence: $$X(\omega) = \sum_{n=-\infty}^{\infty} -jx_R[n]\sin(\omega n)$$

1. Since this is a sum of purely imaginary numbers, $X(\omega)$ is **imaginary**.
1. Also, $X(-\omega) = \sum_{n=-\infty}^{\infty} -jx_R[n]\sin(-\omega n) = \sum_{n=-\infty}^{\infty} jx_R[n]\sin(\omega n) = -X(\omega)$, so it is **odd**.
---

__Answer for 1.4:__

Similar to 1.3, we will be left with only $x_I[n]\sin(\omega n)$. $$X(\omega) = \sum_{n=-\infty}^{\infty} x_I[n]\sin(\omega n)$$

1. Since this is a sum of real numbers, $X(\omega)$ is **real**.
2. Following 1.3, $X(\omega)$ is **odd**.

2. $$x[n] = \begin{cases} 1, \quad &-N\leq n\leq N, \\ 0, &\text{otherwise}. \end{cases}$$ 
   Determine the DTFT for $N = 5, 25, 100$. Scale the
DTFT so that $X(0)=1.$ Comment on the behavior of the magnitude plot as a function of N.

````python
# Code for question 4

def dtft(x, n, w):
    """ DTFT computed at w

    Parameters:
    ---
        x : finite duration sequence over n
        n : sample position vector
        w : frequency location vector

    Returns:
    ---
        X : DTFT values computed at frequencies w
    """
    N = len(n)
    X = np.zeros_like(w)
    for i in range(N):
        X = X + x[i]*np.exp(-1j*n[i]*w)

    mag_X = np.abs(X)
    angle_X = np.angle(X)
    fig, (ax1, ax2) = plt.subplots(2,1, sharex=True)
    ax1.plot(w, mag_X/mag_X.max())
    ax1.set_title('Magnitude')
    ax1.grid()
    ax2.plot(w, angle_X)
    ax2.set_title('Phase')
    ax2.set_xlim(-np.pi, np.pi)
    ax2.grid()

    return X

x_1 = [4,3,2,1,2,3,4]
n = np.arange(7)
w = np.arange(-np.pi, np.pi, 0.05)
X_1 = dtft(x_1, n, w)

K = 5
x_2 = np.ones(2*K)
n = np.arange(-K, K, 1)
X_2 = dtft(x_2, n, w)

K = 25
x_3 = np.ones(2*K)
n = np.arange(-K, K, 1)
X_3 = dtft(x_3, n, w)

K = 100
x_4 = np.ones(2*K)
n = np.arange(-K, K, 1)
X_4 = dtft(x_4, n, w)
````
