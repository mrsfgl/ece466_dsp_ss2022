```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5, 5, 0.5)
y = np.sin(x)
plt.plot(x, y)
plt.plot([0,0], [-2,2], color = 'gray')
plt.plot([-2,2], [0,0], color = 'gray')
plt.xlim(-2,2)
plt.grid()
plt.xlabel('Time (s)')
plt.ylabel('sin(t)')
plt.savefig('Analog.png')
```

### Digital

Digital should be on integer indices, i.e. $n\in N$.

```python
x = np.arange(-5, 5, 0.5)
y = np.sin(x)
x = np.arange(-x.size/2, x.size/2, 1)
plt.stem(x, y, basefmt=' ')
plt.plot([0,0], [-2,2], color = 'gray')
plt.plot([-10,10], [0,0], color = 'gray')
plt.xlim(-x.size/2+1,x.size/2-1)
plt.grid()
plt.xlabel('n (n)')
plt.ylabel('sin(n)')
plt.savefig('Digital.png')
```
