```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
```

```python
n = np.arange(100, dtype=float)
# x = np.zeros_like(n)
x = (-1)**n
b = [1, 1]
a = [1, 1/4, -1/8]
ic = [4, -2]

z = signal.lfiltic(b, a, ic)
y = signal.lfilter(b, a, x, zi=z)
y_2 = (7/12)*(1/4)**n-(5/6)*(-1/2)**n

# print(y)
plt.stem(y[0], basefmt=' ', label='y[n]', linefmt='*r')
plt.stem(y_2, basefmt=' ', linefmt= '--g', label='y_t[n]')
# plt.stem(x, basefmt=' ', label='x[n]', linefmt='--c')
plt.legend()
plt.show
```

```python
y = signal.lfilter(b, a, x)
y_zsr = (1/3)*(1/4)**n+(2/3)*(-1/2)**n

plt.stem(y, basefmt=' ', label='y[n]', linefmt='*r')
plt.stem(y_zsr, basefmt=' ', label='$y+{ZSR}[n]$', linefmt='--g')
# plt.stem(x, basefmt=' ', label='x[n]', linefmt='--c')
plt.legend()
plt.show
```

$H(z)=\frac{0.866}{1-0.8z^{-1}+0.64z^{-2}}=\frac{0.866z^2}{z^2-0.8z^{1}+0.64}$

```python
b = [0.866, 0, 0]
a = [1, -0.8, 0.64]

lti_system = signal.dlti(b,a,dt=1)
_, h = signal.dimpulse(lti_system, t = np.arange(100))

# print(h[0])
plt.stem(h[0].flatten())
```

```python
u = np.ones(150)
y = np.convolve(u, h[0].flatten())

plt.stem(y[:150])
```
