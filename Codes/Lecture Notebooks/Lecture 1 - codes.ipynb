$x[n] = \sin(2\pi fn)$

Digital signals can be expressed as:
1. Formulas, functions.
2. Tabular.
3. Sequence: $x[n] = \{..., 0.02, 0.13, 0.3,...\}$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = 0.05
n = np.arange(-100, 100, 1)
x = np.sin(2*np.pi*f*n)

plt.stem(n, x)
plt.xlim(-10, 30)
plt.xlabel('n')
plt.ylabel('x[n]')
```

__Dirac delta__:
$$ \delta[n] =
\begin{cases}
1, \; n = 0 \\
0, \; \text{oth.}
\end{cases} $$

__Unit step__:
$$ u[n] = 
\begin{cases}
 1, \; n \geq 0 \\
 0, \; n < 0
\end{cases}$$

$$ u[n] - u[n-1] = \delta[n] $$

```python
u = n>=0
plt.stem(n,u)
plt.xlim(-10,10)
```

__Unit ramp__:
$$ r[n] = n u[n]$$

```python
r = n*u
plt.stem(n, r)
plt.xlim(-10, 10)
plt.ylim(-5, 20)
plt.grid()
```

__Exponential signal__:
$$ y[n] = a^n$$

```python
y = np.double(2)**n
plt.semilogy(n, y)
```

__Complex exponential__:
$$a = r^n e^{j\theta n} = r^n \cos(\theta n) + j r^n \sin(\theta n)$$

```python
r = np.double(0.9) # 1.1
theta = 60

a = r**n* np.exp(theta*n*1j)
plt.plot(n,np.abs(a))
```

```python
plt.stem(n, np.angle(a))
plt.xlim(-10, 10)
```
